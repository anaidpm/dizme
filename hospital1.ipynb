```python
import pandas as pd
df = pd.read_csv('data\HHC_SOCRATA_HHCAHPS_PRVDR.csv', encoding = "ISO-8859-1")
df.head()
```

```python
df.columns.values
```

```python
df.set_index('CMS Certification Number (CCN)', inplace=True)
```

```python
df.head()
```

```python
#Drop footnotes
to_drop = []
for column in df:
    if column.startswith("Footnote"):
        to_drop.append(column)
df = df.drop(to_drop, axis=1)
```

```python
df = df.drop(['Provider Name', 'Address', 'City', 'Zip', 'Phone'], axis = 1)
df.head()
```

```python
answers = ['No', 'Yes']
def answer_to_numeric(x):
    for index, r in enumerate(answers):
        if x==r:
            return index
    return -1
```

```python
def ownership_to_numeric(x):
    ownership_types = pd.unique(df['Type of Ownership'].values.ravel())
    for index, r in enumerate(ownership_types):
        if x==r:
            return index
```

```python
states = pd.unique(df['State'].values.ravel())
def state_to_numeric(x):
    for index, r in enumerate(states):
        if x==r:
            return index
```

```python
import datetime
now = datetime.datetime.now()
def extract_years_certification(x):
    years = int(x[:4])
    return now.year - years
```

```python
new_df = pd.DataFrame()
offersColumns = ['Offers Nursing Care Services', 'Offers Physical Therapy Services',
       'Offers Occupational Therapy Services',
       'Offers Speech Pathology Services',
       'Offers Medical Social Services',
       'Offers Home Health Aide Services']

for column in offersColumns:
    new_df[column]= df[column].apply(answer_to_numeric)

new_df.head()
```

```python
for column in df:
    if column not in offersColumns:
        new_df[column]= df[column]

new_df.head()
```

```python
new_df['Type of Ownership']= df['Type of Ownership'].apply(ownership_to_numeric)
new_df.head()
```

```python
new_df['Years Certified'] = new_df['Date Certified'].apply(extract_years_certification)
new_df = new_df.drop(['Date Certified'], axis = 1)
new_df.head()
```

```python
# Replace "not available"
new_df = new_df.replace('Not Available', '-1')
new_df.head()
```

```python
new_df['State']= new_df['State'].apply(state_to_numeric)
new_df.head()
```

```python
#Calculate mutual information, showing which features influence the rating most
from sklearn.feature_selection import mutual_info_classif

feature_scores = mutual_info_classif(new_df, new_df["HHCAHPS Survey Summary Star Rating"].astype(int))
```

```python
import numpy as np
norm_feature_scores = feature_scores/np.max(feature_scores)
features = []
fscores = []
for score, fname in sorted(zip(norm_feature_scores, new_df.columns.values), reverse=True)[:30]:
    print(fname, score)
    features.append(fname)
    fscores.append(score)
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['figure.figsize'] = (10, 3)
rcParams['figure.dpi'] = 100
x = range(len(features))
plt.xticks(x, features, rotation=90)

plt.bar(x,fscores, align='center', alpha=0.7)
plt.show()
```

```python
# Import libraries
import pandas as pd
import folium

# Load the shape of the zone (US states)
# Find the original file here: https://github.com/python-visualization/folium/tree/master/examples/data
# You have to download this file and set the directory where you saved it
state_geo = 'us-states.json'
state_data = state_avg
 
# Initialize the map:
m = folium.Map(location=[37, -102], zoom_start=4)
 
# Add the color for the chloropleth:
m.choropleth(
 geo_data=state_geo,
 name='choropleth',
 data=state_data,
 columns=['State', 'HHCAHPS Survey Summary Star Rating'],
 key_on='feature.id',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.2,
 legend_name='Overall Ratings by State'
)
folium.LayerControl().add_to(m)
m
```
